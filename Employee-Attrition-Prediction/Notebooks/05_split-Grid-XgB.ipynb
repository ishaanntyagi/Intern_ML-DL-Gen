```python
import pandas as pd 
import numpy as np 
```

```python
df = pd.read_csv(r"C:\Users\ishaan.narayan\Desktop\Ishaan's Workspace\Intern_ML-DL-Gen\Employee-Attrition-Prediction\Data\cleaned_employee_data.csv")
print(df.head())
```

```python
df['Attrition'] = df['Attrition'].map({'Yes': 1, 'No': 0})
X = df.drop('Attrition',axis=1)
y = df['Attrition']
# If your target column is called 'Attrition', do this before train/test split:

print(X.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.2)
print(X_train.shape)
print(X_test.shape)
print(y_test.shape)
print(y_train.shape)
```

```python
my_param_grid = {
    'n_estimators': [100, 200, 300],  # Reduced range, more common values
    'max_depth': [3, 5, 7, 10],       # More realistic max_depth for this dataset size
    'learning_rate': [0.01, 0.1, 0.2],
    'subsample': [0.8, 1.0]          # Common good values
}
    
```

```python
from sklearn.model_selection import GridSearchCV
%pip install xgboost
import xgboost as xgb
xgb_model = xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
```

```python
grid_search_xgb = GridSearchCV(
    estimator=xgb_model,
    param_grid=my_param_grid,
    n_jobs=-1,
    scoring='accuracy',    
)
```

```python
grid_search_xgb.fit(X_train,y_train)
```

```python
best_parameters = grid_search_xgb.best_params_
print(best_parameters)
```

```python
print("Best Parameters:", grid_search_xgb.best_params_)
print("Best Cross-Validation Score:", grid_search_xgb.best_score_)

best_model_final = grid_search_xgb.best_estimator_
print(grid_search_xgb.best_estimator_)

from  sklearn.metrics import accuracy_score


y_predicted = best_model_final.predict(X_test)
print(accuracy_score(y_predicted, y_test))
```

```python
import joblib
joblib.dump(best_model_final, 'best_model_XGB.pkl')
```
