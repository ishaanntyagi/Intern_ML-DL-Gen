```python
%pip install seaborn
import seaborn as sns 
import pandas as pd 
import matplotlib.pyplot as plt 
# print(pd)

```

```python
#loading a dataset to work on the sns lib 
df = sns.load_dataset("tips")
print(df.head)
print(df.columns)
```

```python
#using the seaborn for the scatter plot 
sns.scatterplot(data=df , x='total_bill', y='tip', hue='sex', style='time', palette='Set2')
plt.xlabel('TotalBill')
plt.ylabel('Tips')
plt.grid(True)
plt.show()
```

```python
#making a barPlot

sns.barplot(data=df , x='day', y='tip', estimator='mean', palette='Set1')
plt.title("barplot")
plt.grid(True)
plt.show() 

 
```

```python
#sns.boxplot 

sns.boxplot(data=df, x='day', y='tip', palette='Set2')
```

```python
sns.violinplot(data=df, x='day', y='tip', palette='Set2')
```

```python
a=sns.pairplot(df)
b=sns.pairplot(df[['total_bill', 'time']])
plt.show(a)
plt.show(b)
```

```python
#now using HeatMap 
#Firstly calclulating the CorrelationMatrix
#df.corr ( a panda function --> will be used for the correlation matric calculation)
corr = df.corr(numeric_only=True)
sns.heatmap(corr , annot=True , cmap = 'coolwarm', linewidth=0.5, linecolor='black')
plt.grid(True)
plt.show()
```

```python
#catplot
sns.catplot(data=df, x='day', y='tip', kind='box', hue='sex')
```

```python
#now applying Facetgrid --> basically a library in Seaborn

g = sns.FacetGrid(df, col='time', row='day')
g.map(sns.scatterplot, 'total_bill', 'tip')
# plt.show()
```
