```python
import matplotlib.pyplot as plt
x=[1,2,3]
y=[4,5,90]
plt.plot(x,y , color ='red',linestyle = '-', marker='s')
plt.title("MY PLT")
plt.xlabel("X AXIS")
plt.ylabel("Y AXIS")
plt.show()
```

```python
fig, axs = plt.subplots(2, 2) #used for subplots 

# Now axs is a 2D array of Axes objects
axs[0, 0].plot([1, 2, 3], [4, 5, 6])
axs[1, 1].hist([1, 2, 2, 3, 3, 3, 3, 3, 3, 4, 4, 5])
axs[0, 1].scatter([1, 2, 3], [4, 5, 6])
axs[1, 0].bar([1, 2, 3], [4, 5, 6])
plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
#usage of legend box in the top of the plot
x  = [1,2,3,4,5]
y1 = [1,4,9,16,25]
y2 = [2,3,4,5,6]

plt.plot(x,y1,color='blue',linestyle='--',marker='s', label='ABC')
plt.plot(x,y2,color='red',linestyle='-',marker='o', label='EFG')
plt.legend()
plt.show()

```

```python
#axis Example 
x=[1,2,3,4,5,]
y=[10,20,30,40,50]
z=[2,4,6,8,10]

plt.plot(x,y,color='blue',linestyle='--',marker='s', label='x*10')
plt.plot(x,z,color='black',linestyle='-',marker='o', label='x*2')
plt.xlim(2,5)
plt.ylim(0,50)
plt.grid(True)
plt.legend()
plt.show()

```

```python
#bar chart 

Class = ['A','B','C','D']
Values = [1,2,3,4]

plt.bar(Class,Values,color='yellow')
```

```python
#histogram
data = [1,2,2,3,3,3,4,4,4,4,5,5,6,6,7]
plt.hist(data,bins=6,color='green', edgecolor='blue')
```

```python
#scatter Plots 

X01 = [1, 2, 3, 4, 5]
Y01 = [2, 4, 1, 8, 7]
plt.scatter(X01,
            Y01,
            color='red',
            marker="o",
            s=10
            )
plt.show()




new_x1 = [1, 2, 3, 4, 5]
new_y1 = [2, 4, 1, 8, 7]

new_x2 = [1, 2, 3, 4, 5]
new_y2 = [1, 3, 2, 6, 5]

plt.scatter(new_x1, new_y1, color='blue', marker = 's', label='Group A',edgecolors='black')
plt.scatter(new_x2, new_y2, color='red', marker='x', label='Group B',edgecolors='black')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#pieChart (x, labels)
ABC = ['Apples', 'Bananas', 'orranges', ]
sizes = [33,33,57] 

plt.pie(sizes, labels=ABC, startangle=90, normalize=True)
plt.show()
```

```python
import matplotlib
matplotlib.use('TkAgg')

import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

fig, ax = plt.subplots()
line, = ax.plot([], [], 'ro-')

ax.set_xlim(0, 10)
ax.set_ylim(0, 100)
x_data, y_data = [], []

def update(frame):
    x_data.append(frame)
    y_data.append(frame**2)
    line.set_data(x_data, y_data)
    return line,

ani = FuncAnimation(fig, update, frames=range(11), interval=500)
plt.show()
```
