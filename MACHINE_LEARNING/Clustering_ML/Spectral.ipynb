```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import SpectralClustering
import networkx as nx
from sklearn.neighbors import kneighbors_graph
```

```python
X , Y = make_moons(n_samples=1700, noise=2, random_state=45)
print(X.shape, Y.shape)
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
#.fit for mean and std , 
#Scaling -->> transform
#passing together in one function alike element.
```

```python
#moon Vsualization
plt.figure(figsize=(6, 4))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], color='blue', edgecolor='red')
plt.grid(True)
plt.show()
# plt.close()
```

```python
#lets Perfrom Spectral for the Sample Data now 
spectral_new  =  SpectralClustering(
    n_clusters = 2,     
    affinity='nearest_neighbors', 
    n_neighbors=10, 
    random_state=42
    )

labels = spectral_new.fit_predict(X_scaled)
print(labels)

```

```python
plt.figure(figsize=(8, 6))
# plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=labels, cmap='virdis', edgecolor='Blue')
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=labels, cmap='plasma', edgecolor='black')
plt.grid(True)
plt.show()
```

```python
#gpt Provided code to create a KNN graph and visualize it........
from sklearn.neighbors import kneighbors_graph
import networkx as nx
import matplotlib.pyplot as plt

# Step 1: Create KNN graph
knn_graph = kneighbors_graph(X_scaled, n_neighbors=10, include_self=False) 


G = nx.from_scipy_sparse_array(knn_graph)

# Step 3: Draw graph with node positions
plt.figure(figsize=(4, 3))
nx.draw(G, pos={i: X_scaled[i] for i in range(len(X_scaled))}, 
        node_size=20, 
        edge_color='black', 
        node_color='purple')
plt.grid(True)
plt.show()
```
